```python
import random

def load_words(file_path):
    words = {}
    with open(file_path, 'r') as file:
        for word in file:
            word = word.strip().upper()  
            length = len(word)
            if length not in words:
                words[length] = []
            words[length].append(word)  
    return words

def select_word(words, word_length):
    return random.choice(words[word_length]) if word_length in words else None  

def play_game():
    word_length = int(input("Enter the length of the word: "))
    words = load_words("words.txt")
    solution = select_word(words, word_length)
    
    if not solution:
        print("No words found with that length.")
        return

    attempt = 7
    display = "_ " * len(solution)
    letters_found = ""
    score=0

    while attempt > 0:
        print("\nGuess the word: ", display)
        proposition = input("Guess a letter: ")[0:1].upper()

        if proposition in letters_found:
            print("You've already guessed that letter.")
            continue

        letters_found += proposition

        if proposition in solution:
            print("-> Good guess!")
        else:
            attempt -= 1
            print ("wrong guess/n")
            if attempt==0:
                print(" ==========Y= ")
            if attempt<=1:
                print(" ||/       |  ")
            if attempt<=2:
                print(" ||        0  ")
            if attempt<=3:
                print(" ||       /|\ ")
            if attempt<=4:
                print(" ||       /|  ")
            if attempt<=5:                    
                print("/||           ")
            if attempt<=6:
                print("==============\n")

        display = ""
        for x in solution:
            if x in letters_found:
                display += x + " "
                score+=10
            else:
                display += "_ "
                score-=5

        if "_" not in display:
            print("You won!")
            break

    print("\n End of the game ", solution, "score:", score)


if __name__ == "__main__":
    play_game()
```
